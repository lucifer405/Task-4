```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("globalterrorismdb_0718dist.csv")

data
```

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','attacktype1_txt':'AttackType','targtype1_txt':'TargetType','gname':'Group','weaptype1_txt':'Weapon','nkill':'Kill','nwound':'Wound'},inplace=True)
```

```python
data=data[['Year','Month','Day','Country','Region','State','City','latitude','longitude','summary','AttackType','TargetType','Group','Weapon','Kill','Wound','target1','motive']]
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Wound']=data['Wound'].fillna(0)
data['Kill']=data['Kill'].fillna(0)
```

```python
data['casualities']=data['Wound']+data['Kill']
```

```python
data.info()
```

```python
data.describe()
```

```python
year=data['Year'].unique()
years_count=data['Year'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(18,10))
sns.barplot(x=year,y=years_count)
plt.xticks(rotation=90)
plt.xlabel('YEARS' , fontsize=20)
plt.ylabel('No. Of Attacks Each Year' , fontsize=20)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

```python
pd.crosstab(data.Year , data.Region).plot(kind='hist',stacked=False,figsize=(20,10))
plt.xlabel('Years',fontsize=20)
plt.ylabel('No. of Attacks' , fontsize=20)
plt.title('Terrorist By Region In Each Year' , fontsize=30)
plt.show()
```

```python
attack=data.Country.value_counts()[:10]
attack
```

```python
Groups_attacks=data.Group.value_counts()[1:10]
Groups_attacks
```

```python
plt.subplots(figsize=(18,10))
sns.barplot(data['Country'].value_counts()[:10].index , data['Country'].value_counts()[:10].values)
plt.xlabel('Countries' , fontsize=20)
plt.ylabel('Count' , fontsize=20)
plt.title('Top Attacked Countries',fontsize=30)
plt.show()
```

```python
df=data[['Year','Kill']].groupby(['Year']).sum()
fig ,ax4=plt.subplots(figsize=(18,10))
df.plot(kind='bar',ax=ax4)
plt.xticks(rotation=90)
plt.xlabel('Years',fontsize=20)
plt.ylabel('No. of Killed Peoples',fontsize=20)
plt.title('Total Died Due To Attack',fontsize=30)
plt.show()
```

```python
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(18,10),color='green')
plt.xlabel('Name of City',fontsize=20)
plt.ylabel('No. Of Attacks',fontsize=20)
plt.title('Top 10 Attacked Cities',fontsize=30)
plt.show()
```

```python
data['AttackType'].value_counts().plot(kind='bar',color='red',figsize=(18,10))
plt.xlabel('Types of Attacks',fontsize=20)
plt.ylabel('No. Of Attacks',fontsize=20)
plt.title('AttackType Evaluation',fontsize=30)
plt.show()
```

```python
data[['AttackType','Kill']].groupby(['AttackType'],axis=0).sum().plot(figsize=(18,10),color='violet',kind='bar')
plt.xlabel('Type Of Attack',fontsize=20)
plt.ylabel('No. Of People',fontsize=20)
plt.title('Total Attacked People',fontsize=30)
plt.show()
```

```python
data[['AttackType','Wound']].groupby(['AttackType'],axis=0).sum().plot(figsize=(18,10),color='yellow',kind='bar')
plt.xlabel('Types Of Attacks',fontsize=20)
plt.ylabel('No. Of People',fontsize=20)
plt.title('Total No. Of Wounded',fontsize=30)
plt.show()
```

```python
plt.subplots(figsize=(18,10))
sns.countplot(data['TargetType'],order=data['TargetType'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Types Of Targets',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.title('Target Type Visualization',fontsize=30)
plt.show()
```

```python
data['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='cyan',figsize=(18,10))
```

```python
data[['Group','Kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values(['Kill'],ascending=False).head(10).plot(figsize=(18,10),color='orange',kind='bar')
plt.xlabel('Terrorist Group Name',fontsize=20)
plt.ylabel('Number of Killed People',fontsize=20)
plt.title('Top 10 Terrorist Group Attack',fontsize=30)
plt.show()
```

```python
df=data[['Group','Country','Kill']]
df=df.groupby(['Group','Country'],axis=0).sum().sort_values('Kill',ascending=False).drop('Unknown').reset_index().head(10)
df
```
